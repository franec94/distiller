```python
sol_files = []
def get_files(target):
    sol = []
    for dirName, subdirList, fileList in os.walk(target, topdown=False):
        # print('Found directory: %s' % dirName)
        for fname in fileList:
            if fname == 'model_final.pth':
                sol.append(os.path.join(dirName, fname))
                pass
            # print('\t%s' % fname)
            pass
        pass
    return  sol
for row in df_occr.values[:]:
    target = row[-1]
    list_files = get_files(target)
    if list_files is []: continue
    n_attempts = len(list_files) // row[3]
    for trial in range(row[3]):
        arch_no = f"arch_no_{trial}"
        for attempt in range(n_attempts):
            attemp_no = f'trial_no_{attempt}'
            merged = os.path.join(arch_no, attemp_no)
            # print(merged)
            for a_file in list_files:
                if merged in a_file:
                    # print(a_file)
                    sol_files.append(a_file)
                    pass
                pass
            pass
        pass
    pass
cropped_images_df['path'] = sol_files
cropped_images_df
```

```python
def read_csv_data(a_file_csv):
    # - Read data from src file
    runs_df = pd.read_csv(a_file_csv)

    if 'Unnamed: 0' in runs_df.columns:
        runs_df = runs_df.drop(['Unnamed: 0'], axis = 1)
        pass
    cropped_images_df = runs_df[~runs_df['cropped_heigth'].isna()][~runs_df['cropped_width'].isna()]

    # - Sort data
    attrs_for_sorting = "timestamp,hf,hl".split(",")
    cropped_images_df = cropped_images_df.sort_values(by = attrs_for_sorting)

    return cropped_images_df
```

```python
columns = cropped_images_df.columns
RowTuple = collections.namedtuple('RowTuple', columns)
def create_path(row, split_val = '\\', a_root = '\\content\\my_data'):
    a_row_tuple = RowTuple._make(row)
    
    path = a_row_tuple.path
    arch_no, trial_no = path.split(f'{split_val}')[6:8]
    file_name = list(map(str,[a_row_tuple.timestamp, a_row_tuple.hf, a_row_tuple.hl, arch_no, trial_no, 'pth']))
    file_name = '.'.join(file_name)
    path = [a_root, file_name]
    a_path = os.path.join(*path)
    a_path = a_path.replace("\\", "/")
    return a_path
```
